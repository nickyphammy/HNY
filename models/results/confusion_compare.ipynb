## Comparing False Positive and False Negatives

```python
import json
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# ============ Loading files ==============
with open('word2vec_spellcorrected_confusion_matrix.txt', 'r') as f:
    word2vec_spell = json.load(f)
with open('word2vec_sub_confusion_matrix.txt', 'r') as f:
    word2vec = json.load(f)
```

```python
# =========== Find common false positives and false negatives ==============
word2vec_spell_false_pos = word2vec_spell['False Positives']
word2vec_spell_false_neg = word2vec_spell['False Negatives']
word2vec_false_pos = word2vec['False Positives']
word2vec_false_neg = word2vec['False Negatives']

both_false_pos = set(word2vec_spell_false_pos).intersection(set(word2vec_false_pos))
both_false_neg = set(word2vec_spell_false_neg).intersection(set(word2vec_false_neg))
spell_only_false_pos = set(word2vec_spell_false_pos) - set(word2vec_false_pos)
spell_only_false_neg = set(word2vec_spell_false_neg) - set(word2vec_false_neg)
norm_only_false_pos = set(word2vec_false_pos) - set(word2vec_spell_false_pos)
norm_only_false_neg = set(word2vec_false_neg) - set(word2vec_spell_false_neg)

print(f"Common false positives: {len(both_false_pos)}")
print(f"Common false negatives: {len(both_false_neg)}")
print(f"Spell-corrected false positives only: {len(spell_only_false_pos)}")
print(f"Spell-corrected false negatives only: {len(spell_only_false_neg)}")
print(f"Normal false positives only: {len(norm_only_false_pos)}")
print(f"Normal false negatives only: {len(norm_only_false_neg)}")
```

```python
# ============ Save the results ==============
results = {
    "Common False Positives": list(both_false_pos),
    "Common False Negatives": list(both_false_neg),
    "Spell-corrected False Positives Only": list(spell_only_false_pos),
    "Spell-corrected False Negatives Only": list(spell_only_false_neg),
    "Normal False Positives Only": list(norm_only_false_pos),
    "Normal False Negatives Only": list(norm_only_false_neg)
}
with open('word2vec_comparison.json', 'w') as f:
    json.dump(results, f, indent=4)
```

## Further Exploration on Cleaned Mendeley and Kaggle Data (Written by Nicholas)

```python
import pandas as pd
import json
import os
import re
import matplotlib.pyplot as plt
```

```python
# ============ LOAD DATA ============
print("Loading data...")

# Load Mendeley data
with open('../cleaned_data/mendeley_cleaned.json', 'r', encoding='utf-8') as f:
    mendeley = json.load(f)

# Load Kaggle data
with open('../cleaned_data/kaggle_data.json', 'r', encoding='utf-8') as f:
    kaggle = json.load(f)

print(f"Mendeley reviews: {len(mendeley)}")
print(f"Kaggle reviews: {len(kaggle)}")
```

```python
# ============ PLOT DISTRIBUTION OF MENDELEY REVIEW LENGTHS BY LABEL ============
m_pos_df = pd.DataFrame([review for review in mendeley if review['recommend'] == 'Recommended'])
m_neg_df = pd.DataFrame([review for review in mendeley if review['recommend'] == 'Not Recommended'])
m_pos_lens = m_pos_df['review'].str.len()
m_pos_avg_len = m_pos_lens.mean()
m_neg_lens = m_neg_df['review'].str.len()
m_neg_avg_len = m_neg_lens.mean()
print(f"Average length of recommended reviews: {m_pos_avg_len:.2f} characters")
print(f"Average length of not recommended reviews: {m_neg_avg_len:.2f} characters")
plt.figure(figsize=(10, 6))
plt.hist([m_pos_lens, m_neg_lens], bins=80, alpha=0.8, label=['Recommended', 'Not Recommended'], color=['cornflowerblue', 'indianred'])
plt.yscale('log')
plt.xlabel('Review Length (characters)')
plt.ylabel('Number of Reviews (log scale)')
plt.legend()
plt.title('Distribution of Review Lengths by Label')
plt.show()
```

```python
# ============ PLOT DISTRIBUTION OF KAGGLE REVIEW LENGTHS BY LABEL ============
k_pos_df = pd.DataFrame([review for review in kaggle if review['is_positive'] == 1])
k_neg_df = pd.DataFrame([review for review in kaggle if review['is_positive'] == 0])
k_pos_lens = k_pos_df['review_content'].str.len()
k_pos_avg_len = k_pos_lens.mean()
k_neg_lens = k_neg_df['review_content'].str.len()
k_neg_avg_len = k_neg_lens.mean()
print(f"Average length of recommended reviews: {k_pos_avg_len:.2f} characters")
print(f"Average length of not recommended reviews: {k_neg_avg_len:.2f} characters")
plt.figure(figsize=(10, 6))
plt.hist([k_pos_lens, k_neg_lens], bins=80, alpha=0.8, label=['Recommended', 'Not Recommended'], color=['cornflowerblue', 'indianred'])
plt.yscale('log')
plt.xlabel('Review Length (characters)')
plt.ylabel('Number of Reviews (log scale)')
plt.legend()
plt.title('Distribution of Review Lengths by Label')
plt.show()
```
